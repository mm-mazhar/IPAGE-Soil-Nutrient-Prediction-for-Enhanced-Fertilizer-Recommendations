# Corralation with outliers

```python
from sklearn.preprocessing import LabelEncoder

for col in df.select_dtypes(include=['object']).columns:
    encoder = LabelEncoder()
    df[col] = encoder.fit_transform(df[col].astype(str))
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True, cmap="coolwarm",fmt=".2f")
plt.title("Corralation with outliers")
plt.show()
```

# Corralation without outliers

```python
conditions = {
    'Zinc Zn (ug/g)': df['Zinc Zn (ug/g)'] > 2.5,
    'Boron B (ug/g)': df['Boron B (ug/g)'] > 0.800,
    'Sulfur S (ug/g)': df['Sulfur S (ug/g)'] > 32,
    'Phosphorus P (ug/g)': df['Phosphorus P (ug/g)'] > 80,
    'Potassium K (meq/100)': df['Potassium K (meq/100)'] > 0.600,
    'Nitrogen N (%)': df['Nitrogen N (%)'] > 0.25,
    'SOC (%)': df['SOC (%)'] > 3.75,
    'pH': (df['pH'] < 4.2) | (df['pH'] > 6.5)
}

grouped = df.groupby(['Area', 'Data Collection Year', 'Land class'])
```

```python
for column, condition in conditions.items():
    median_values = grouped[column].transform('median')
    df.loc[condition, column] = median_values[condition]
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True, cmap="coolwarm",fmt=".2f")
plt.title("Corralation without outliers")
plt.show()
```

```python
for col in df.columns:
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    sns.histplot(df[col], ax=axes[0])
    sns.boxplot(x=df[col], ax=axes[1])
    plt.title(f"Distribution of {col} without outliers")
    plt.tight_layout()
    for ax in axes:
        ax.tick_params(axis='x', rotation=90)
    plt.show()
    print('\n')
```
