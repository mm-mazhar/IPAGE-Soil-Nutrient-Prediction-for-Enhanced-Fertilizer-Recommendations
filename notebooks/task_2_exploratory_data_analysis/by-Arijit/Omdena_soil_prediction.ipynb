Display basic information about the dataset

```python
import pandas as pd

# Load the data
file_path = '/content/2410_iPAGE_SoilData_cleaned.csv'
data = pd.read_csv(file_path)

data_info = data.info()
data_head = data.head()

data_info, data_head
```

```python
# Check for missing values
missing_values = data.isnull().sum()

# Statistical summary of numerical columns
statistics_summary = data.describe()

# Count of unique values in categorical columns
categorical_summary = data.select_dtypes(include=['object']).nunique()

# Select only numeric columns for the correlation matrix
correlation_matrix = data.select_dtypes(include=['float64', 'int64']).corr()

# Display the correlation matrix
correlation_matrix

# Output summaries
missing_values, statistics_summary, categorical_summary, correlation_matrix
```

Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

data.hist(bins=15, figsize=(15, 10))
plt.suptitle('Data Distributions')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(data.select_dtypes(include=['float64', 'int64']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
